## Ejemplo de uso

Red neuronal sencilla, asigna pesos y umbrales alueatorios si no son dados

## Ejemplo 2

Genera Red neuronal con datos especificos

```python
# Ejemplo de uso
layers = np.array([3, 4, 2])
net2 = net.NeuralNetwork(layers)

# Asignar pesos desde un diccionario
weights_dict = {
    (0, 3): 0.5, (0, 4): 0.3, (0, 5): -0.2, (0, 6): 0.8,
    (1, 3): -0.6, (1, 4): 0.1, (1, 5): 0.7, (1, 6): -0.4,
    (2, 3): 0.2, (2, 4): -0.9, (2, 5): 0.3, (2, 6): 0.5,
    (3, 7): 0.4, (3, 8): -0.3,
    (4, 7): 0.1, (4, 8): 0.6,
    (5, 7): -0.7, (5, 8): 0.2,
    (6, 7): 0.9, (6, 8): -0.5
}
net2.weights(weights_dict)

# Asignar umbrales desde un diccionario
thresholds_dict = {
    3: -0.3, 
    4: 0.5, 
    5: 0.1, 
    6: -0.7, 
    7: 0.2, 
    8: -0.4
}
net2.thresholds(thresholds_dict)

# Definir la función de activación sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

net2.set_activation_function(sigmoid)

# Visualizar la red neuronal con pesos y umbrales
net2.plot()
```

```python
inputs = np.array([1,2,4])

output = net2.evaluate(inputs)
print(output)
```
